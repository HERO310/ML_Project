# ICE-LoRA Loss Function
### Implement ICE and combined loss functions

```
import torch
import torch.nn.functional as F

def ice_loss(logits_context, logits_query):
    p_cq = F.log_softmax(logits_context, dim=-1)
    p_q = F.softmax(logits_query, dim=-1)
    kl = F.kl_div(p_cq, p_q, reduction='batchmean')
    return kl

def total_loss(ft_loss, ice, lam=0.5):
    return ft_loss + lam * ice

# Example usage with dummy tensors
logits_context = torch.randn(2, 10)
logits_query = torch.randn(2, 10)
ft = F.cross_entropy(logits_context, torch.randint(0, 10, (2,)))
ice = ice_loss(logits_context, logits_query)
print('Total Loss:', total_loss(ft, ice))
```
